## DAPPER During Train

```python
base_result_dir = os.path.join(main_result_dir, "from_embeds", "train_dapper")
levels = ["user", "seed", "alpha"]
dapper_train_results = ResultProcessor(base_result_dir, levels, verbose=True)
```

```python
per_alpha_results = dapper_train_results.results_df.groupby(["alpha", "user"]).agg(["mean", "std"])[["eval_loss", "eval_accuracy"]]
per_alpha_results
```

```python
# get mean and standard of loss and accuracy for each alpha
per_alpha_results.groupby("alpha").agg(["mean", "std"])
```

```python
# find best alpha for each user based on validation perf
per_user_results = dapper_train_results.results_df.groupby(["user", "alpha"]).agg(["mean", "std"])[["eval_loss", "eval_accuracy", "test_loss", "test_accuracy"]]
per_user_results
```

```python
# how much does performance vary by alpha within users
var_across_alpha = per_user_results.groupby('user').std()
var_across_alpha
```

```python
var_across_alpha.mean()
```

```python
var_across_alpha.std()
```

```python
min_idx = per_user_results.groupby("user").transform(min)[("eval_loss", "mean")] == per_user_results[("eval_loss", "mean")]
min_idx
```

```python
max_idx = per_user_results.groupby("user").transform(max)[("eval_loss", "mean")] == per_user_results[("eval_loss", "mean")]
max_idx
```

```python
worst_alpha_df = per_user_results.loc[max_idx]
worst_alpha_df = worst_alpha_df.reset_index()
worst_alpha_df["alpha"] = worst_alpha_df["alpha"].apply(pd.to_numeric)
worst_alpha_df
```

```python
best_alpha_df = per_user_results.loc[min_idx]
best_alpha_df = best_alpha_df.reset_index()
best_alpha_df["alpha"] = best_alpha_df["alpha"].apply(pd.to_numeric)
best_alpha_df
```

```python
# get diff in best and worst alphas
diff = best_alpha_df[("test_accuracy", "mean")] - worst_alpha_df[("test_accuracy", "mean")]
print(diff.mean())
print(diff.std())
```

```python
# get diff in best and worst alphas
diff = best_alpha_df[("test_loss", "mean")] - worst_alpha_df[("test_loss", "mean")]
print(diff.mean())
print(diff.std())
```

```python
best_alpha_df.agg(["mean", "std"])
```

```python
best_alpha_df.groupby("alpha").count()
```

```python
best_alpha_df.quantile(q=[0, .2, .4, .6, .8, 1])
```
