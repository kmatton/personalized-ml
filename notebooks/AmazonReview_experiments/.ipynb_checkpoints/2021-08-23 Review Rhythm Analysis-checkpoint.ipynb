## Load Data

```python
data_dir = "/data/ddmg/redditlanguagemodeling/data/AmazonReviews/data"
```

```python
data_df = pd.read_csv(os.path.join(data_dir, 'amazon_v2.0/reviews.csv'),
                      dtype={'reviewerID':str, 'asin':str, 'reviewTime':str,'unixReviewTime':int,
                             'reviewText':str,'summary':str,'verified':bool,'category':str, 'reviewYear':int},
                      keep_default_na=False, na_values=[], quoting=csv.QUOTE_NONNUMERIC)
```

```python
data_df = data_df.rename(columns={"reviewText": "text"})
```

```python
data_df = data_df.rename(columns={"overall": "score"})
```

```python
split_df = pd.read_csv(os.path.join(data_dir, 'amazon_v2.0', 'splits', 'user.csv'))
```

```python
data_df = data_df[split_df["split"] != -1]
```

```python
len(data_df)
```

# Get Review Scores w/ Per-User Z-Normalization

```python
data_df['index'] = data_df.index
```

```python
data_df.head(10)
```

```python
def z_score(x, mean, std):
    if std:
        return (x - mean) / std
    return np.nan
```

```python
user_mean_scores = data_df.groupby('reviewerID').agg('mean')['score']
user_std_scores = data_df.groupby('reviewerID').agg('std')['score']
```

```python
user_mean_scores
```

```python
user_std_scores
```

```python
data_df['z_norm_score'] = data_df.apply(lambda x: z_score(x['score'], user_mean_scores[x['reviewerID']], user_std_scores[x['reviewerID']]), axis=1)
```

```python
cleaned_df = data_df.dropna(subset=['z_norm_score'])
```

```python
len(cleaned_df)
```

```python
len(cleaned_df.reviewerID.unique())
```

```python
cleaned_df['reviewTime'] = pd.to_datetime(cleaned_df['reviewTime'])
```

```python
cleaned_df["dayofweek"] = cleaned_df['reviewTime'].dt.dayofweek
```

```python
cleaned_df["is_weekend"] = cleaned_df['dayofweek'].isin([5,6])
```

```python
cleaned_df['reviewYear'] = cleaned_df['reviewYear'].astype(int)
```

```python
cleaned_df['reviewMonth'] = cleaned_df['reviewTime'].dt.month
```
