## Get Global Model Preds on Data Subset (to test result processing code)

```python
base_result_dir = os.path.join(main_result_dir, "from_embeds", "eval_global_model_on_same_data")
levels = ["train_seed", "eval_seed"]
global_results = ResultProcessor(base_result_dir, levels, verbose=True)
```

```python
def get_max_class(x):
    return np.argmax(x[["pred_cls_1", "pred_cls_2", "pred_cls_3", "pred_cls_4", "pred_cls_5"]]) + 1
```

```python
# add in final preds
global_results.preds_df["pred"] = global_results.preds_df.apply(lambda x: get_max_class(x), axis=1)
```

```python
global_results.preds_df.groupby(["sample_idx"]).agg(["mean", "std"])
```

```python
global_results.preds_df.groupby(["sample_idx"]).agg(["mean", "std"])["pred", "std"].mean()
```

## Get Person-Specific Model Preds on Subset of Data

```python
base_result_dir = os.path.join(main_result_dir, "from_embeds", "eval_person_specific_my_split_n_500_on_same_data")
levels = ["user", "train_seed", "eval_seed"]
ps_results = ResultProcessor(base_result_dir, levels, verbose=True)
```

```python
# check performance
# much worse than performance of global model, which is expected
ps_results.results_df.agg(["mean", "std"])[["test_accuracy", "test_loss"]]
```

```python
# get standard deviation in performance for each user across seeds
ps_results.results_df.groupby("user").agg(["mean", "std"])[["test_loss", "test_accuracy"]]
```

```python
ps_results.results_df.groupby("user").agg(["mean", "std"])[["test_loss", "test_accuracy"]].mean()
```

```python
ps_results.results_df.groupby("user").agg(["mean", "std"])[["test_loss", "test_accuracy"]].std()
```

```python
# add in final preds
ps_results.preds_df["pred"] = ps_results.preds_df.apply(lambda x: get_max_class(x), axis=1)
```

```python
ps_results.preds_df.sort_values(["sample_idx", "user"])
```

```python
# note: standard deviation of per-user models is much higher across different seeds
ps_user_results = ps_results.preds_df.groupby(["user", "sample_idx"]).agg(["mean", "std"])
ps_user_results
```

```python
# mean(standard deviation in preds across seeds) across users
print(ps_user_results["pred", "std"].mean())
print(ps_user_results["pred", "std"].std())
```

```python
# compare standard deviation for pred for given sample *within* users (+ across seeds) to standard deviation across users
# if standard deviation within users is similar or greater than across users, this analysis won't make sense
user_mean_preds = ps_results.preds_df.groupby(["sample_idx", "user"]).agg(["mean", "std"])
user_mean_preds
```

```python
# for each sample, get standard deviation of mean pred across users
sample_user_mean_preds = user_mean_preds.groupby(["sample_idx"]).agg(["mean", "std"])
sample_user_mean_preds
```

```python
print(sample_user_mean_preds['pred', 'mean', 'std'].mean())
print(sample_user_mean_preds['pred', 'mean', 'std'].std())
```
