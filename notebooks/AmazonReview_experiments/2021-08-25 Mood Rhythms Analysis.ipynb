## Load Data

```python
data_dir = "/data/ddmg/redditlanguagemodeling/data/AmazonReviews/data"
```

```python
data_df = pd.read_csv(os.path.join(data_dir, 'amazon_v2.0/reviews.csv'),
                      dtype={'reviewerID':str, 'asin':str, 'reviewTime':str,'unixReviewTime':int,
                             'reviewText':str,'summary':str,'verified':bool,'category':str, 'reviewYear':int},
                      keep_default_na=False, na_values=[], quoting=csv.QUOTE_NONNUMERIC)
```

```python
data_df = data_df.rename(columns={"reviewText": "text", "overall": "score"})
```

```python
split_df = pd.read_csv(os.path.join(data_dir, 'amazon_v2.0', 'splits', 'user.csv'))
data_df = data_df[split_df["split"] != -1]
len(data_df)
```

## Add Sentiment Scores

```python
nltk.download('vader_lexicon')
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
# try it out
text_ex = data_df.text.values[0]
text_ex
```

```python
# how are scores computed?
# 1. get valance of each word
# 2. to get compound scores --> get sum of valances for each word --> normalize by: score / sqrt(score^2 + alpha)
#     alpha approximates max expected value (they use alpha = 15)
# 3. for each word, shift score as follows: if positive add 1, if negative add -1 + keep track of counts of 0 scores
# 4. normalize pos, neg, neutral to sum to 1
```

```python
# note on how compound score is computed: https://stackoverflow.com/questions/40325980/how-is-the-vader-compound-polarity-score-calculated-in-python-nltk
sia.polarity_scores(text_ex)
```

```python
data_df["sentiment"] = data_df["text"].apply(lambda x: sia.polarity_scores(x))
```

```python
for feat in ['neg', 'neu', 'pos', 'compound']:
    data_df["sentiment_{}".format(feat)] = data_df["sentiment"].apply(lambda x: x[feat])
```

```python
def z_score(x, mean, std):
    if std:
        return (x - mean) / std
    return np.nan
```

```python
# get sentiment normalized per-user
for feat in ['neg', 'neu', 'pos', 'compound']:
    user_mean_scores = data_df.groupby('reviewerID').agg('mean')['sentiment_{}'.format(feat)]
    user_std_scores = data_df.groupby('reviewerID').agg('std')['sentiment_{}'.format(feat)]
    data_df['z_norm_sentiment_{}'.format(feat)] = data_df.apply(lambda x: z_score(x['sentiment_{}'.format(feat)], 
                                                                                  user_mean_scores[x['reviewerID']],
                                                                                  user_std_scores[x['reviewerID']]
                                                                                  ), axis=1)
```
