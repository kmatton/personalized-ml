```python
# look at difference
perc_of_og = []
srs = ["AskDocs", "Bitcoin", "investing", "mentalhealth", "relationship_advice", "relationships"]
for sr in srs:
    og_count = len(val_df[val_df['subreddit'] == sr])
    new_count = len(new_val[new_val['subreddit'] == sr])
    try:
        ratio = new_count/og_count
    except:
        print(sr)
    perc_of_og.append(ratio * 100)
```
