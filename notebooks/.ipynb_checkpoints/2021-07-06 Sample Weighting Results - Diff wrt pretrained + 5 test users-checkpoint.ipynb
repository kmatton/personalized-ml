```python
def collect_result(result_dir, pretrained_perf_df=None, test_users=None):
    full_dir = os.path.join(base_result_dir, result_dir)
    ea = ExpAnalyzer(full_dir)
    if test_users is not None:
        ea.author_result_df = ea.author_result_df[ea.author_result_df["author"].isin(test_users)]
    print("Mean Perplexity Per Author: {}".format(ea.author_result_df.mean()["perplexity"]))
    print("Std Perplexity Per Author: {}".format(ea.author_result_df.std()["perplexity"]))
    if pretrained_perf_df is not None:
        merged_df = pretrained_perf_df.merge(ea.author_result_df, how="inner", on="author")
        merged_df["perplexity_diff"] = merged_df.apply(lambda x: x['perplexity_x'] - x['perplexity_y'], axis=1)
        print("Mean Perplexity Diff Per Author: {}".format(merged_df.mean()["perplexity_diff"]))
        print("Std Perplexity Diff Per Author: {}".format(merged_df.std()["perplexity_diff"]))
        return merged_df
    return ea.author_result_df
```

```python
def collect_per_user_result(result_dir, pretrained_perf_df=None, test_users=None):
    full_dir = os.path.join(base_result_dir, result_dir)
    ea = PerPopulationExpAnalyzer(full_dir, "author")
    author_result_df = ea.data_split_perf_df[ea.data_split_perf_df["data_split"] == "test"]
    if test_users is not None:
        author_result_df = author_result_df[author_result_df["author"].isin(test_users)]
    print("Mean Perplexity Per Author: {}".format(author_result_df.mean()["perplexity"]))
    print("Std Perplexity Per Author: {}".format(author_result_df.std()["perplexity"]))
    if pretrained_perf_df is not None:
        merged_df = pretrained_perf_df.merge(author_result_df, how="inner", on="author")
        merged_df["perplexity_diff"] = merged_df.apply(lambda x: x['perplexity_x'] - x['perplexity_y'], axis=1)
        print("Mean Perplexity Diff Per Author: {}".format(merged_df.mean()["perplexity_diff"]))
        print("Std Perplexity Diff Per Author: {}".format(merged_df.std()["perplexity_diff"]))
        return merged_df
    return author_result_df
```

```python
def collect_per_sample_sr_result(result_dir, pretrained_perf_df=None, test_users=None):
    full_dir = os.path.join(base_result_dir, result_dir)
    ea = PerPopulationExpAnalyzer(full_dir, "subreddit")
    sample_df = ea.sample_losses_df.merge(test_df, on="id", how="inner")
    author_result_df = sample_df.groupby("author").mean()["loss"].to_frame()
    author_result_df["perplexity"] = author_result_df["loss"].apply(lambda x: loss_to_perplexity(x))
    author_result_df = author_result_df.reset_index()
    if test_users is not None:
        author_result_df = author_result_df[author_result_df["author"].isin(test_users)]
    print("Mean Perplexity Per Author: {}".format(author_result_df.mean()["perplexity"]))
    print("Std Perplexity Per Author: {}".format(author_result_df.std()["perplexity"]))
    if pretrained_perf_df is not None:
        merged_df = pretrained_perf_df.merge(author_result_df, how="inner", on="author")
        merged_df["perplexity_diff"] = merged_df.apply(lambda x: x['perplexity_x'] - x['perplexity_y'], axis=1)
        print("Mean Perplexity Diff Per Author: {}".format(merged_df.mean()["perplexity_diff"]))
        print("Std Perplexity Diff Per Author: {}".format(merged_df.std()["perplexity_diff"]))
        return merged_df
    return author_result_df
```
